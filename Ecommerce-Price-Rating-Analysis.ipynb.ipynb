```python
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "/content/drive/MyDrive/Product_Comparison_Amazon_Flipkart_Myntra (2).xlsx"
df = pd.read_excel(file_path)
```

```python
print("\nDataFrame columns immediately after loading:")
print(df.columns.tolist())
```

```python
for col in ['Amazon ₹', 'Flipkart ₹', 'Myntra ₹']:
    if col in df.columns:
        df[col] = (
            df[col].astype(str)
            .str.replace('₹', '', regex=False)
            .str.replace(',', '', regex=False)
            .str.strip()
        )
        df[col] = pd.to_numeric(df[col], errors='coerce')
    else:
        print(f"Column '{col}' not found in the DataFrame. Skipping currency conversion.")
```

```python
if 'Ratings (A/F/M)' in df.columns:
    # Ensure the column is treated as string to safely use .str accessor
    df['Ratings (A/F/M)'] = df['Ratings (A/F/M)'].astype(str)
    rating_split = df['Ratings (A/F/M)'].str.split('/', expand=True)
    rating_split = df['Ratings (A/F/M)'].str.split('/', expand=True)
```

```python
if 'Ratings (A/F/M)' in df.columns:
    df['Ratings (A/F/M)'] = df['Ratings (A/F/M)'].astype(str)
    rating_split = df['Ratings (A/F/M)'].str.split('/', expand=True)

    if rating_split.shape[1] > 0:
        df['Rating_A'] = pd.to_numeric(rating_split[0].str.strip(), errors='coerce')
    if rating_split.shape[1] > 1:
        df['Rating_F'] = pd.to_numeric(rating_split[1].str.strip(), errors='coerce')
    if rating_split.shape[1] > 2:
        df['Rating_M'] = pd.to_numeric(rating_split[2].str.strip(), errors='coerce')
else:
    print("Ratings (A/F/M) column not found.")
```

```python
required_columns = ['Product', 'Amazon ₹', 'Flipkart ₹', 'Myntra ₹', 'Rating_A', 'Rating_F', 'Rating_M']

if all(col in df.columns for col in required_columns):
    long_df = pd.DataFrame({
        'Product': df['Product'].tolist() * 3,
        'Platform': ['Amazon'] * len(df) + ['Flipkart'] * len(df) + ['Myntra'] * len(df),
        'Price': pd.concat([df['Amazon ₹'], df['Flipkart ₹'], df['Myntra ₹']], ignore_index=True),
        'Rating': pd.concat([df['Rating_A'], df['Rating_F'], df['Rating_M']], ignore_index=True)
    })
     # Add category if it exists
    if 'Product Category' in df.columns:
        long_df['Category'] = df['Product Category'].tolist() * 3

    long_df.dropna(subset=['Price', 'Rating'], inplace=True)

    # Display final transformed DataFrame
    print("\nTransformed DataFrame (long format):")
    display(long_df.head())
else:
    print("One or more required columns are missing. Cannot reshape the DataFrame.")

```

```python
long_df.to_csv("product_comparison_long.csv", index=False)
```
